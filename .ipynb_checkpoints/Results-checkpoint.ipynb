```python
import glob
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
results_path = './results/*'
results = glob.glob(results_path)
print(results)
```

```python
result_n = 3

with open(results[result_n], 'rb') as handle:
    result = pickle.load(handle)
```

```python
def plot_results(res, idx):
    print("Result dimensions are: {}".format(res.shape))
    plt.figure(figsize=(30,6))
    for i in range(res.shape[-1]):
        plt.subplot(2, 10, i+1)
        plt.imshow(res[idx, :, :, i], cmap='gray')
        plt.axis('off')
        
def plot_all_results(res):
    print("Result dimensions are: {}".format(res.shape))
    idx = 0
    for j in range(res.shape[0]):
        plt.figure(figsize=(30,6))
        for i in range(res.shape[-1]):
            plt.subplot(2, 10, i+1)
            plt.imshow(res[idx, :, :, i], cmap='gray')
            plt.axis('off')
        idx += 1
        print("========================================================================")
```

```python
plot_results(result, 0)
```

```python
plot_results(result, 1)
```

```python
plot_all_results(result,1)
```
