```python
import glob
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
results_path = './results/*'
results = glob.glob(results_path)
```

```python
result_n = 0

with open(results[result_n], 'rb') as handle:
    result = pickle.load(handle)
```

```python
def plot_results(res, idx):
    print("Result dimensions are: {}".format(res.shape))
    plt.figure(figsize=(30,6))
    for i in range(res.shape[-1]):
        plt.subplot(2, 10, i+1)
        plt.imshow(res[idx, :, :, i], cmap='gray')
        plt.axis('off')
```

```python
plot_results(result, 7)
```
